```python
import gymnasium as gym
import matplotlib.pyplot as plt
import time

#create mountain car environment
env=gym.make("MountainCar-v0",render_mode="rgb_array")

#reset environment + return (state,info)
state,info=env.reset()

print("\nInitial state (position,velocity):",state)
print("Number of position actions:",env.action_space.n) #3 actions
```

```python
#for storing rewards
episodes=200
reward_sum=0
for i in range(episodes):
  #Render current frame
  frame=env.render()
  plt.imshow(frame)
  plt.axis('off')
  plt.pause(0.001) #shows animation smoothly
  plt.clf() #clear frame

  #Take random action
  action=env.action_space.sample()
  print(f"Step {i} -> Action Taken:", action)

  #Steps return
  state,reward,done,truncated,info=env.step(action)
  reward_sum+=reward

  #If car reaches end or max steps hits - restart episode
  if done or truncated:
    print("\n Episode Finished!")
    print("-> Total Reward Collected:",reward_sum)
    print("-> Resetting Environment_ _ _ \n")
env.close()
```

```python
import gymnasium as gym
import matplotlib.pyplot as plt
env=gym.make("MountainCar-v0",render_mode="rgb_array")
state,info=env.reset()
print("Controls: A = Left | S = Neutral | D = Right")
for step in range(2):
  frame=env.render()
  plt.imshow(frame)
  plt.axis('off')
  plt.pause(0.01) #shows animation smoothly
  plt.clf() #clear frame

  #User Input
  move = input("Enter Movement (A/S/D): ").lower()
  if move == 'a':
    action = 0 #left
  elif move == 's':
    action = 1 #neutral(no push)
  elif move == 'd':
    action = 2 #right
  else:
    print("Invalid choice! Choose only A/S/D")
    continue
  state,reward,done,truncated,info=env.step(action)

  #If goal reached
  if state[0] >= 0.5:
    print("\n You have reached the goal successfully! You have climbed the hill!")
    break
env.close()
```
