```python
#Creating a list
numbers=[2,4,6,8,10]
colors=["Red","blue","pink","yellow","black"]
combined_list=[109,80.9,"Ashifa",True]
empty_list=[]
print(numbers,colors,combined_list,empty_list)
```

```python
#Accessing list elements
colors=["Red","blue","pink","yellow","black"]
print(colors[0])
print(colors[-5])
print(colors[4])
print(colors[-1])
```

```python
#Modify a list
colors=["Red","blue","pink","yellow","black"]
colors[1]="Lavender"
colors[2]="Green"
print(colors)
```

```python
#Common list operations-01
colors=["Red","blue","pink","yellow","black"]
colors.append("White")
colors.insert(1,"Orange")
print(colors)
```

```python
#Common list operations-02
colors=["Red","blue","pink","yellow","black"]
colors.pop()
print(colors)
```

```python
#Common list operations-04
colors=["Red","blue","pink","yellow","black"]
combined_list=[109,80.9,"Ashifa",True]
print(len(colors))
print(len(combined_list))
```

```python
#Common list operations-03
colors=["Red","blue","pink","yellow","black"]
colors.remove("Red")
print(colors)
```

```python
#Build-in functions in list
a=[5,34,98,20,25]
print(min(a))
print(max(a))
a.sort()
print(a)
a.reverse()
print(a)
```
