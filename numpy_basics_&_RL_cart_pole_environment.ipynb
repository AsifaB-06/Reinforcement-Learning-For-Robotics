#Importing Libraries

```python
import gymnasium as gym
import numpy as np
import time
```

#Setting the environment

```python
env=gym.make('CartPole-v1',render_mode='rgb_array')
```

#Resetting the environment (INITIAL STATE) & Moving the cart pole using step()

```python
#prints the initial state of the cart pole(cart position,cart velocity,pole angle,angular velocity)
state=env.reset()
print(state)
#Moving the cart to left
state=env.step(0)
print(state)
#Moving the cart to right
state=env.step(1)
print(state)
```

#Displaying the cart pole using matplotlib

```python
import matplotlib.pyplot as plt
state=env.unwrapped.state #To bring for initial state
print(state) #printing statement for shape
frame=env.render() #render() is used for visualizing
plt.imshow(frame) #imshow() - image show
plt.axis('off') #To turn off the axis(x,y)
plt.show()
```

#Changing the cart movement // Change the position of the cart [0] th index of the state

#Observation space (storing the the previous observation in a variable called obs_space

```python
obs_space=env.observation_space
```

#Extract the high and low limits (similarly storing the values in a variable)

```python
high=obs_space.high
low=obs_space.low
```

#Pole angle is the 3rd value (index 2)

```python
max_angle_rad=high[2]
min_angle_rad=low[2]
```

#Convert to degrees for easier understanding

```python
max_angle_deg=np.degrees(max_angle_rad)
min_angle_deg=np.degrees(min_angle_rad)
print("==== Pole Angle Limits ====")
print(f"Min Angle(radians):{min_angle_rad}")
print(f"Max Angle(radians):{max_angle_rad}")
print()
print(f"Min Angle(degrees):{min_angle_deg:.2f}")
print(f"Max Angle(degrees):{max_angle_deg:.2f}")
```

#FULL CART POLE PROJECT

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import time
env=gym.make("CartPole-v1",render_mode='rgb_array')
epoisodeNumber=5
timeSteps=100
for episode in range(epoisodeNumber):
  result=env.reset()#To start from initial
  for t in range(timeSteps):
    frame=env.render()
    frame=frame[0] #will take the entire array of state values
    frame=np.array(frame)
    plt.imshow(frame)
    plt.title(f"Episode{episode},Step{t}")
    plt.axis('off')
    plt.show()
    #Take Random action
    step_result=env.step(env.action_space.sample())
    #Handle gym old/new versions(4 or 5 outputs)
    if len(step_result)==5:
      obs,reward,terminated,truncated,info=step_result
    else:
      obs,reward,done,info=step_result
      terminated,truncated=done,False
      #slow animation(increase for slower)
    time.sleep(0.1)
    if terminated or truncated:
      break
  env.close()
```
